### Importación de librerías

```python
# pip install -r requirements.txt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

# Análisis exploratorio de datos y preprocesamiento

### Carga de datos

```python
data = pd.read_csv('heart_statlog_cleveland_hungary_final.csv')
data
```

### Correlación de variables

```python
sns.heatmap(data.corr(), annot=True)
```

### Distribución de variables

```python
sns.pairplot(data=data)
```

### Preprocesamiento

```python
data = data.dropna()
x_data = data.iloc[:, 1:-1].values
y_data = data.iloc[:, -1].values
x_data_train, x_data_test, y_data_train, y_data_test = train_test_split(x_data, y_data, test_size=0.3, random_state=0)
x_data_train_scaled = StandardScaler().fit_transform(x_data_train)
x_data_test = StandardScaler().fit_transform(x_data_test)

y_data_train = y_data_train.reshape(-1, 1)
```

```python
print(y_data_train)
```

## Arquitectura 1

### Creación del modelo

```python
model = tf.keras.models.Sequential()
# Capa 1
model.add(tf.keras.layers.Dense(units=10, activation='relu'))

# Capa 2
model.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Capa de salida
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy', tf.keras.metrics.Recall()])
```

### Summary del Modelo

```python
model.summary()
```

```python
model.fit(x_data_train_scaled, y_data_train, batch_size=8, epochs=200, validation_data=(x_data_test, y_data_test))
```

Como el accuracy y el val_acuracy se parecen, esto quiere decir que no hay overfitting. Asimismo, el val_acurracy se mantiene constante durante el tiempo.

## Arquitectura 2

### One Hot Encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ct2 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
y_data_encoded_2 = np.array(ct2.fit_transform(y_data_train))

print(y_data_encoded_2)
```

### Creación del modelo

```python
model2 = tf.keras.models.Sequential()

# Capa 1
model2.add(tf.keras.layers.Dense(units=10, activation='relu'))

# Capa 2
model2.add(tf.keras.layers.Dense(units=6, activation='relu'))

# Capa de salida
model2.add(tf.keras.layers.Dense(units=2, activation='softmax'))

model2.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', tf.keras.metrics.Recall()])
```

### Summary del Modelo

```python
model2.summary()
```

```python
model2.fit(x_data_train_scaled, y_data_encoded_2, batch_size=8, epochs=100)
```

## Arquitectura 3

### Creación del modelo

```python
model3 = tf.keras.models.Sequential()

# Capa 1
model3.add(tf.keras.layers.Dense(units=5, activation='relu'))

# Capa 2
model3.add(tf.keras.layers.Dense(units=15, activation='relu'))

# Capa 3
model3.add(tf.keras.layers.Dense(units=10, activation='relu'))

# Capa 4
model3.add(tf.keras.layers.Dense(units=5, activation='relu'))

# Capa de salida
model3.add(tf.keras.layers.Dense(units=2, activation='softmax'))

model3.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy', tf.keras.metrics.Recall()])
```

### Summary del Modelo

```python
model3.fit(x_data_train_scaled, y_data_encoded_2, batch_size=8, epochs=50)
```
